```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("nasdaq100.csv",sep=";")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["Date"])
```

```python
plt.scatter(df["Starting (USD)"],df["Ending (USD)"], color='red')
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.show()
```

```python
x=df[["Starting (USD)"]]
x.head()
```

```python
y=df[["Ending (USD)"]]
y.head()
```

**Linear Regression**

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(), color='green')
plt.scatter(df["Starting (USD)"],df["Ending (USD)"], color='red')
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.show()
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
m=reg.coef_
c=reg.intercept_
```

```python
print(m*16700+c)
```

```python
reg.predict([[16700]])
```

```python
df['prediced_y']=reg.predict(x)
df.head()
```

```python
plt.plot(df["Starting (USD)"],df["prediced_y"],color='blue')
plt.scatter(x.mean(),y.mean(), color='green')
plt.scatter(df["Starting (USD)"],df["Ending (USD)"], color='red')
plt.xlabel("Starting (USD)")
plt.ylabel("Ending (USD)")
plt.show()
```
